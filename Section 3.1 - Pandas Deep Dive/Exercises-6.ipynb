### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
# First combine these sales together into a single dataframe, then compute 
# how much money consumers spent on each book in each currency

sales1 = sales1.rename(columns={'Book title': 'Title', 
                                'Number sold': 'Units sold', 
                                'Sales price': 'List price', 
                                'Royalty paid': 'Royalty'})
sales_new = pd.concat([sales1, sales2])

col = sales_new.Title.str.extract(r'\((.*)\)')
sales_new['currency'] = col.fillna(method='bfill')

sales_new['Total'] = sales_new['Units sold'] * sales_new['List price']
sales_new = sales_new[sales_new['Total'].notnull()]
sales_new.groupby(['Title','currency'])[['Total']].sum()
```
